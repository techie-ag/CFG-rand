```python
#Import libraries

import pandas as pd
import requests
import pprint as pp
```

```python
#Import borough data into dataframe

df = pd.read_csv(r'C:\Users\augus\CFG4-Yelp\Beth - working docs\yelp_ealing.csv')
df.head(20)

#Set new FHRS columns

df['FHRSID'] = ''
df['FHRSBusinessName'] = ''
df['FHRSPostcode'] = ''
df['FHRSRating'] = ''

df
```

```python
# Loop through dataframe to make FHRS API call for each entry

for i in df.index:
    name = df['name_x'][i]
    longitude = df['longitude'][i]
    latitude = df['latitude'][i]
    maxDistanceLimit = 1
    url = f'https://api.ratings.food.gov.uk/Establishments?name={name}&longitude={longitude}&latitude={latitude}&maxDistanceLimit={maxDistanceLimit}'
    headers = {'x-api-version': '2', 'accept': 'application/json'}
    resp = requests.get(url, headers = headers).json()
    if resp['establishments'] == []:
        FHRSID = 'N/A'
        FHRSID = 'N/A'
        FHRSBusinessName = 'N/A'
        FHRSPostcode = 'N/A'
        FHRSRating = 'N/A'
    else:
        FHRSID = resp['establishments'][0]['FHRSID']
        FHRSBusinessName = resp['establishments'][0]['BusinessName']
        FHRSPostcode = resp['establishments'][0]['PostCode']
        FHRSRating = resp['establishments'][0]['RatingValue']
        
    df['FHRSID'][i] = FHRSID
    df['FHRSBusinessName'][i] = FHRSBusinessName
    df['FHRSPostcode'][i] = FHRSPostcode
    df['FHRSRating'][i] = FHRSRating
```

```python
FHRS_ratings = df['FHRSRating']

print(FHRS_ratings.value_counts())
```

```python
# # Check data match

for i in df.index:
    if df['name_x'][i] != df['FHRSBusinessName'][i]:
        if df['FHRSBusinessName'][i] != 'N/A':
            print(df['name_x'][i], ' - ', df['FHRSBusinessName'][i])
```
